## WORKING ON IRIS DATASET

This toy data set contains measuremnet of sepal length, sepal width, petal lenght an petal width of 3 different species of irs flowers

**GOAL**: Tak at hand is to use the sepal features in kmeans clustering of the three species i.e.  *setosa*,*veriscolor* and *virgincia*

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing and loading the dataset
from sklearn.datasets import load_iris
iris=load_iris(as_frame=True)
#splitting to feature and target
X=iris.data
y=iris.target
```

```python
X.head()
```

VISUALIZATION USING SEPAL DETAILS


```python

fig, ax = plt.subplots(figsize=(6, 4))

# Plot the data points using only the sepal information
scatter = ax.scatter(iris.data['sepal length (cm)'], iris.data['sepal width (cm)'], c=y, cmap= 'viridis')

# Add labels
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_title('Iris Dataset (Sepal)')

#Add legend
ax.legend( scatter.legend_elements()[0], iris.target_names, loc="lower right")

plt.show()
```

CLUSTERING USING KMEANS

```python
# visualizing the data pointS as they are
fig,ax=plt.subplots(figsize=(8,6))
# line above is to create a figure and an axis
ax.scatter(X['sepal length (cm)'],X['sepal width (cm)'],color='blue')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Iris Dataset')
plt.show()
```

```python
# now clustering
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,random_state=42,n_init='auto')
labels=km.fit_predict(X)
labels
```

```python
# viewing our centroids
centroids=km.cluster_centers_
centroids
```

```python
#visualizing the clustering
plt.figure(figsize=(6,4))

plt.scatter(X['sepal length (cm)'],X['sepal width (cm)'],c=labels, cmap='viridis', alpha=0.5, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=300, c='r', label='Centroids')

plt.show()
```
